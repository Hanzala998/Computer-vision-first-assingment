```python
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    tokens = word_tokenize(text)
    stop_words = set(stopwords.words('english'))
    tokens_no_stopwords = [word for word in tokens if word not in stop_words]
    stemmer = PorterStemmer()
    stemmed_tokens = [stemmer.stem(word) for word in tokens_no_stopwords]
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(word) for word in tokens_no_stopwords]
    return {
        "original_text": text,
        "tokens": tokens,
        "tokens_no_stopwords": tokens_no_stopwords,
        "stemmed_tokens": stemmed_tokens,
        "lemmatized_tokens": lemmatized_tokens
    }
```

```python
from google.colab import files
print("Please upload a text file:")
uploaded = files.upload()
```

```python
for filename in uploaded.keys():
    with open(filename, 'r', encoding='utf-8') as file:
        input_text = file.read()
```

```python
results = preprocess_text(input_text)
```

```python
output_filename = "2021-SE-05_Assignment_01_NLP_Preprocessed_Results.txt"
with open(output_filename, 'w', encoding='utf-8') as output_file:
    for step, output in results.items():
        print(f"\n{step.replace('_', ' ').capitalize()}:\n{output}")
        output_file.write(f"{step.replace('_', ' ').capitalize()}:\n{output}\n\n")

print(f"\nAll results have been saved to {output_filename}.")
```
